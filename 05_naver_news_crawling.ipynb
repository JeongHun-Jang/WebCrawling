# Naver Newsnow Title Crawling

### 이 시각 주요뉴스 10개 크롤링

```python
# input : url 
# return: bs4.BeautifulSoup = soup
def requestPage(url):
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text, 'lxml')
    return soup
```

```python
# input : soup (1 page)
# return: list of news title
def onePageData(soup):
    ul_tag = soup.select('ul.newsnow_txarea')
    
    news_title_list = []
    news_href_list = []
    news_contents_list = []
    
    for tag in ul_tag:
        # news title
        strong_tag_list = tag.select('li > div.newsnow_tx_inner > a > strong')
        for strong_tag in strong_tag_list:
            news_title_list.append(strong_tag.text.strip())
        
        # news href and contents
        a_tag_list = tag.select('li > div.newsnow_tx_inner > a')
        for a_tag in a_tag_list:
            href = a_tag.attrs['href']
            news_href_list.append(href) # append href to list
            news_soup = requestPage(href) # request href
            content = news_soup.find('div',id='articleBodyContents')
            if content == None:
                news_contents_list.append('There are many news in href')
            else:
                # append to list after removing useless string
                str_contents = content.text.strip()
                str_contents = str_contents.replace('// flash 오류를 우회하기 위한 함수 추가\nfunction _flash_removeCallback() {}\n\n','')
                str_contents = str_contents.replace('동영상 뉴스','').strip()
                news_contents_list.append(str_contents)
    # for end            
    news = zip(news_title_list,news_href_list,news_contents_list)
    df_news = pd.DataFrame(news)
    return df_news
```

```python
import bs4
import requests
import pandas as pd
```

```python
site = 'https://news.naver.com/'
soup = requestPage(site)
```

```python
news_df = onePageData(soup)
```
