## 1 page crawling

```python
import bs4
import requests
site = 'https://pjt3591oo.github.io'

response = requests.get(site)

# 응답결과 문자열을 bs4
soup = bs4.BeautifulSoup(response.text,'lxml')
urls = soup.select('body > main div div.p > h3 > a')

print(urls[0].text)
print(urls[0].attrs['href'])
```

```python
div_list = soup.select('div.p')

for div_tag in div_list:
    h3_tag = div_tag.find('h3')
    #print(h3_tag)
    
    a_tag = h3_tag.find('a')
    href = a_tag.attrs['href'] # 태그 속성 가져오기
    post_title = a_tag.text
    print(post_title)
    
    h4_tag = div_tag.find('h4')
    post_desc = h4_tag.text.strip()  # strip() 앞뒤 공백 제거 함수
    print(post_desc, '\n')
```

## all pages crawling 1

using url pattern

```python
import bs4
import requests
import time
```

```python
for page_num in range(4):
    if page_num == 0 :  # page1 -> error code 404
        url = 'https://pjt3591oo.github.io'
    else: # page 2-4
        url = f'https://pjt3591oo.github.io/page{page_num+1}'
    time.sleep(1) # time delay
    
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.text,'lxml')
    div_tag_list = soup.select('body > main div div.p')
    
    for div_tag in div_tag_list:
        a_tag = div_tag.select('h3 > a')
        h4_tag = div_tag.select('h4')
        print('title : ', a_tag[0].text.strip())
        print('desc  : ', h4_tag[0].text.strip())
        print('url   : ', a_tag[0].attrs['href'], '\n')
        
```

## all pages crawling 2

using next page url

```python
import bs4
import requests
import time
```

```python
def getDivList(site):
    response = requests.get(site)
    soup = bs4.BeautifulSoup(response.text,'lxml')
    div_list = soup.find_all('div', class_='p')
    
    # 다음 페이지 확인
    next_tag = soup.find(any, text='Next') # any : 모든 태그
    next_page = None
    if next_tag.name == 'a' :
        next_page = next_tag.attrs['href']
    return div_list, next_page

def getData(div_tag):
    h3_tag = div_tag.find('h3')
    a_tag = h3_tag.find('a')
    href = a_tag.attrs['href']
    a_text = a_tag.text
    
    h4_tag = div_tag.find('h4')
    h4_text = h4_tag.text.strip()
    
    p_tag = div_tag.find('p', class_='author')
    span_tag = p_tag.find('span', class_='date')
    temp_str = span_tag.text.strip()
    temp_list = temp_str.split('|')
    date_str = temp_list[0].strip()
    author_str = temp_list[1].strip()
    
    return href, a_text, h4_text, date_str, author_str

def printList(data_list):
    for dict1 in data_list:
        print('-----------------')
        for key, value in dict1.items():
            
            print(key, " ", value)
        print('-----------------')
```

```python
site = 'https://pjt3591oo.github.io'
div_list = getDivList(site)
dat = []
while True:
    time.sleep(1)    
    div_list, next_page = getDivList(site)
    for a_div in div_list:
        tuple1 = getData(a_div)
        tuple2 = ('href', 'a_text', 'h4_text', 'date_text', 'author')
        zip1 = zip(tuple2, tuple1)
        #print(tuple(zip1))
        dat_dic = dict(zip1)
        dat.append(dat_dic)
    print(printList(dat))
    if (next_page == None):
        break
    else:
        site = f'https://pjt3591oo.github.io/{next_page}'
```
